# 1 Concatenating objects

```python
 df1 = pd.DataFrame({
     'A': ['A0', 'A1', 'A2', 'A3'],
     'B': ['B0', 'B1', 'B2', 'B3'],
     'C': ['C0', 'C1', 'C2', 'C3'],
     'D': ['D0', 'D1', 'D2', 'D3']},
     index=[0, 1, 2, 3])
df1
```

```python
df2 = pd.DataFrame({
    'A': ['A4', 'A5', 'A6', 'A7'],
    'B': ['B4', 'B5', 'B6', 'B7'],
    'C': ['C4', 'C5', 'C6', 'C7'],
    'D': ['D4', 'D5', 'D6', 'D7']},
     index=[4, 5, 6, 7])
df2
```

```python
df3 = pd.DataFrame({
    'A': ['A8', 'A9', 'A10', 'A11'],
    'B': ['B8', 'B9', 'B10', 'B11'],
    'C': ['C8', 'C9', 'C10', 'C11'],
    'D': ['D8', 'D9', 'D10', 'D11']},
     index=[8, 9, 10, 11])
df3
```

```python
frames = [df1, df2, df3]; frames
```

```python
result = pd.concat(frames); result
```

```python
result = pd.concat(frames, keys=['x', 'y', 'z']); result # cria referências para as tabelas mescladas
```

```python
result.loc['y'] 
```

# 2 Databases-style DataFrame joing/merging

## 2.1 Brief primer on merge methods (relationnal algebra)

```python
left = pd.DataFrame({
    'key': ['K0', 'k1', 'K2', 'K3'],
    'A': ['A0', 'A1', 'A2', 'A3'],
    'B': ['B0', 'B1', 'B2', 'B3']
});
left
```

```python
right = pd.DataFrame({
    'key': ['K0', 'k1', 'K2', 'K3'],
    'C': ['C0', 'C1', 'C2', 'C3'],
    'D': ['D0', 'D1', 'D2', 'D3']
});
right
```

```python
result = pd.merge(left, right, on='key'); result # efetuando o INNER JOIN
```

```python
# Exemplo mais complexo com multiindex
left = pd.DataFrame({
    'key1': ['K0', 'K0', 'K1', 'K2'],
    'key2': ['K0', 'K1', 'K0', 'K1'],
    'A': ['A0', 'A1', 'A2', 'A3'],
    'B': ['B0', 'B1', 'B2', 'B3']
});
left
```

```python
right = pd.DataFrame({
    'key1': ['K0', 'K1', 'K1', 'K2'],
    'key2': ['K0', 'K0', 'K0', 'K0'],
    'C': ['C0', 'C1', 'C2', 'C3'],
    'D': ['D0', 'D1', 'D2', 'D3']
});
right
```

```python
result = pd.merge(left, right, on=['key1', 'key2']); result # inner join com duas chaves
```

### 2.1.1 LEFT OUTER JOIN

```python
left
```

```python
right
```

```python
result = pd.merge(left, right, how='left', on=['key1', 'key2']); result # LeFT OUTER JOIN
```

### 2.1.2 RIGHT OUTER JOIN

```python
result = pd.merge(left, right, how='right', on=['key1', 'key2']); result # RIGHT OUTER JOIN
```

### 2.1.3 FULL OUTER JOIN

```python
result = pd.merge(left, right, how='outer', on=['key1', 'key2']); result # FULL OUTER JOIN
```

### 2.1.4 INNER JOIN 

```python
result = pd.merge(left, right, how='inner', on=['key1', 'key2']); result # INNER OUTER JOIN
```

## 2.2 Checking for duplicate keys

```python
left = pd.DataFrame({'A': [1, 2], 'B': [1, 2]})
```

```python
right = pd.DataFrame({'A': [4, 5, 6], 'B': [2, 2, 2]})
```

```python
result = pd.merge(left, right, on='B', how='outer', validate='one_to_one')
# Gerar Exception
# MergeError: Merge keys are not unique in right dataset; not a one-to-one merge
```

```python
pd.merge(left, right, on='B', how='outer', validate='one_to_many')
```

## 2.3 The merge indicator

```python
df1 = pd.DataFrame({'col1': [0, 1], 'col_left': ['a', 'b']})
```

```python
df2 = pd.DataFrame({'col1': [1, 2, 2], 'col_right': [2, 2, 2]})
```

```python
pd.merge(df1, df2, on='col1', how='outer', indicator=True)
```

```python
pd.merge(df1, df2, on='col1', how='outer', indicator='indicator_column')
```

# 3 Joining on index

```python
left = pd.DataFrame({
    'A': ['A0', 'A1', 'A2'],
    'B': ['B0', 'B1', 'B2']},
    index=['K0', 'K1', 'K2'])
left
```

```python
right = pd.DataFrame({
    'C': ['C0', 'C2', 'C3'],
    'D': ['D0', 'D2', 'D3']},
    index=['K0', 'K2', 'K3'])
right
```

```python
result = left.join(right); result
```

```python
result = left.join(right, how='outer'); result
```

```python
result = left.join(right, how='inner'); result
```

## 3.1 Joing key columns on a index

```python
left = pd.DataFrame({
    'A': ['A0', 'A1', 'A2', 'A3'],
    'B': ['B0', 'B1', 'B2', 'B3'],
    'key': ['K0', 'K1', 'K0', 'K1']
});
left
```

```python
right = pd.DataFrame({
    'C': ['C0', 'C1'],
    'D': ['D0', 'D1']
}, index = ['K0', 'K1'])
right
```

```python
result = left.join(right, on ='key'); result
```

```python
result = pd.merge(left, right, left_on='key', right_index=True, how='left', sort=False);
result # Equivalente a instrução anterior
```
