# Concatenar arquivos pelas colunas,  mantendo as quantidades de linhas

```python
col_concat = pd.concat([df1, df2, df3], axis=1) # isso é muito útil
col_concat
```

```python
new_df = col_concat['A'] # ele mantém as referências das colunas dos mesmos nomes e preserva seus valores
new_df
```

```python
col_concat.loc[0]
```

# Trocando os nomes das colunas

```python
new_df.columns = ['A', 'B', 'C'] # É possível renomear as colunas acessando a propriedade "columns". Muito útil
new_df
```

# Trocando os nomes dos índices

```python
new_df.index = ['Y1', 'Y2', 'Y3','Y4'] # muito útil também
new_df
```

# One to One ( site.name = visited_subset.site )

Por default esse merge a seguir é um <strong>INNER</strong>

```python
o2o = pd.merge(site, visited_subset, # tablea left, tabela right
                  left_on='name',
                  right_on='site')
o2o
```

# One to Many

```python
m = site.merge(visited, left_on='name', right_on='site')
m
```

```python
df1 = pd.DataFrame({
    'a': [1, 1, 1, 2, 2, ],
    'b': [10, 20, 30, 40 ,50]
})

df2 = pd.DataFrame({
    'a1': [1, 1, 2, 2, 3],
    'b1': [100, 200, 300, 400, 500]
})
```
