Selecionando a coluna tipo reclamação e retornando as 5 primeiras linhas

```python
complaints['Complaint Type'][:5]
```

Não importa a ordem da chamada das chaves para diferenciar as colunas das linhas da tabela. O resultado será o mesmo:

```python
complaints[:5]['Complaint Type']
```

Contabilizando as ocorrências das reclamações através do método <strong>value_counts()</strong>

```python
complaints['Complaint Type'].value_counts()
```

Podemos armazenar estes dados agrupados com os totais e selecionar os 5 maiores motivos de reclamações

```python
complaint_counts = complaints['Complaint Type'].value_counts()
complaint_counts[:5]
```

Para otimizar a vizualização desses dados, podemos plotar estes registros através de um grafico de barras.

```python
complaint_counts[:5].plot(kind='bar', figsize=[15,5])
```

Podemos plotar este gráfico através do método <strong>bar</strong> da propriedade <strong>plot</strong>

```python
complaint_counts[:5].plot.bar(figsize=[15,5])
```
