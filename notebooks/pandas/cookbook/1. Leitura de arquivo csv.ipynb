# Reading data from a csv file

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
plt.style.use('default')
```

```python
broken_df = pd.read_csv('../data/cookbook/bikes.csv', 
                        sep=';', 
                        encoding='latin1', 
                        parse_dates =['Date'])
```

# Selecting a lines

Selecionando as 3 primeiras linhas do arquivo

```python
broken_df[:3]
```

<div>A listagem da ordem das datas estão com relação aos meses e não em relação aos dias. Para resolver este problema, utilizamos o parâmetro <strong>dayfirst = True</strong></div>

```python
broken_df = pd.read_csv('../data/cookbook/bikes.csv', 
                        sep=';', 
                        encoding='latin1', 
                        parse_dates =['Date'],
                        dayfirst=True)
```

```python
broken_df[:3]
```

Para otimizarmos, vamos definir a coluna <strong>Date</strong> como nosso index

```python
broken_df = pd.read_csv('../data/cookbook/bikes.csv', 
                        sep=';', 
                        encoding='latin1', 
                        parse_dates =['Date'],
                        dayfirst=True,
                       index_col='Date')
```

```python
broken_df[:3]
```

# Selecting a column

Selecionando os dados pela <strong>coluna</strong> e trazendo apenas 3 linhas

```python
broken_df['Berri 1'][:3]
```

Pode ser utilizado também através da propriedade <strong>iloc</strong>

```python
broken_df['Berri 1'].iloc[:3]
```

Pode ser seleciona mais de uma coluna

```python
broken_df[['Berri 1', 'Rachel1']].iloc[:3]
```

# Plotting a column

```python
broken_df['Berri 1'].plot()
```

Ampliando a dimenção da imagem do gráfico

```python
broken_df['Berri 1'].plot(figsize = (15,5))
```
